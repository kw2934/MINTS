```python
m = 100 # total number of independent runs
num_groups = 10 # partition all seeds into this number of groups
id_group = 1 # group index for this notebook. NOTE: Please run the experiment for id_group = 1, 2, ..., num_groups. Use multiple Colab sessions if needed.
file_path = path + 'Experiments/Data/results_pricing_{}_{}.pkl'.format(num_groups, id_group)


# specify the seeds
list_seeds_all = np.arange(1, m + 1) * 100 # list of all random seeds

tmp = m // num_groups
if id_group < num_groups:
    list_seeds = list_seeds_all[(id_group - 1) * tmp : id_group * tmp]
else:
    list_seeds = list_seeds_all[(id_group - 1) * tmp : m]

# candidate prices: 0.05 to 0.95 by 0.05 (19 in total)
K = 19
p_min, p_max = 0.05, 0.95
prices = p_min + (p_max - p_min) * np.arange(K) / (K - 1)

Lipschitz_const = 1 # Lipschitz constant for the constraint
demand_std = 0.5 # standard deviation parameter in the Gaussian likelihood
T = 5000 # number of rounds
list_methods = ['MINTS_pricing_Bernoulli','MINTS_pricing_Gaussian', 'TS', 'UCB1'] # list of methods


# run experiments and save results (decisions of all algorithms; customer valuations at all times)
simulation_dynamic_pricing(T = T, prices = prices, list_seeds = list_seeds, list_methods = list_methods, file_path = file_path, demand_std = demand_std, Lipschitz_const = Lipschitz_const)


```
